# Story Workflow Migration

Adds story workflow tracking columns to `genealogy.silver_person_story_status`.

## New columns
| Column | Type | Description |
|---|---|---|
| `story_stage` | STRING | Workflow stage: `not_started` \| `timeline` \| `context` \| `draft` \| `publish` |
| `timeline_doc_id` | STRING | Google Drive doc ID for the saved timeline research brief |
| `context_doc_id` | STRING | Google Drive doc ID for saved context/Perplexity notes |
| `draft_doc_id` | STRING | Google Drive doc ID for the story draft |

## Design notes
- `story_stage` is set manually from the app — the `/story/stage` PATCH endpoint
  MERGEs into this table, creating a row if absent, so no manual inserts are needed
- All four new columns are nullable — existing rows are unaffected
- Doc IDs are the Google Drive file IDs (stable, do not change on rename)
- Run once; safe to re-run (ADD COLUMNS IF NOT EXISTS not supported in Databricks SQL,
  so wrapped in a TRY block — will error harmlessly if columns already exist)


```python
# Add story_stage column
try:
    spark.sql("""
        ALTER TABLE genealogy.silver_person_story_status
        ADD COLUMN story_stage STRING
    """)
    print("Added: story_stage")
except Exception as e:
    print(f"story_stage already exists or error: {e}")
```
