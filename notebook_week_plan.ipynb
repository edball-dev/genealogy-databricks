# Week Planning — Table Setup

Creates and populates `genealogy.ref_week_plan` used by the Family Tree Research app **This Week** tab.

## Rotation parameters
- **Anchor**: Sunday 8 March 2026 = `week_index` 0
- **Modes** (cycle every week): `housekeeping` → `tree_building` → `dna` → `story_writing`
- **Branches** (cycle every week, independent of mode): `Ball` → `Mattock` → `Cuthbertson` → `McLean` → `Pearson` → `Palmer` → `Easter` → `Clifford`
- **Full non-repeating cycle**: 32 weeks (LCM of 4 modes × 8 branches)
- **Coverage**: 96 rows = 3 × 32-week cycles (2026-03-08 to 2028-01-02)

## Run order
Run cells **1 → 2 → 3** once to set up. Cells 4–6 are ongoing maintenance.


## Cell 1 — Create `ref_week_plan` table

Idempotent — safe to re-run. Will not overwrite existing data.

```python
# Cell 1: Create ref_week_plan table
# No DEFAULT column values — Databricks requires enabling a TBLPROPERTIES setting
# to support column defaults, so all defaults are supplied explicitly in the
# INSERT / MERGE statements in Cell 2 instead.

spark.sql("""
CREATE TABLE IF NOT EXISTS genealogy.ref_week_plan (
  week_commencing  DATE      NOT NULL COMMENT 'Sunday of the week (primary key)',
  week_index       INT       NOT NULL COMMENT 'Rotation position 0-indexed from anchor 2026-03-08',
  mode             STRING    NOT NULL COMMENT 'housekeeping | tree_building | dna | story_writing',
  branch           STRING    NOT NULL COMMENT 'Ball | Mattock | Cuthbertson | McLean | Pearson | Palmer | Easter | Clifford',
  skipped          BOOLEAN   NOT NULL COMMENT 'TRUE = skip this week; app advances to next non-skipped row',
  mode_override    STRING    COMMENT 'NULL = use rotation default. Set to override mode for this week only.',
  branch_override  STRING    COMMENT 'NULL = use rotation default. Set to override branch for this week only.',
  notes            STRING    COMMENT 'Free text e.g. holiday week',
  created_at       TIMESTAMP NOT NULL COMMENT 'Row created or last modified timestamp'
)
USING DELTA
COMMENT '96-week research rotation schedule (3 x 32-week cycles, 2026-03-08 to 2028-01-02).'
""")

print("ref_week_plan table ready.")
```

## Cell 3 — Verify: current week + next 3

Simulates exactly what the app queries on page load.

```python
# Cell 3: Verify the schedule

# Part A: first 36 rows — confirm the rotation pattern is correct
print("=== First 36 rows ===")
display(spark.sql("""
    SELECT week_commencing, week_index, mode, branch,
           skipped, mode_override, branch_override
    FROM   genealogy.ref_week_plan
    ORDER  BY week_commencing
    LIMIT  36
"""))

# Part B: simulate the app query — current week + next 3 non-skipped rows
# DAYOFWEEK: 1=Sunday, 2=Monday ... 7=Saturday
print("\n=== Current week + next 3 (as the app sees it) ===")
display(spark.sql("""
    WITH this_sunday AS (
        SELECT date_sub(current_date(), DAYOFWEEK(current_date()) - 1) AS dt
    ),
    current_and_future AS (
        SELECT
            wp.*,
            wp.week_commencing = (SELECT dt FROM this_sunday) AS is_current_week,
            ROW_NUMBER() OVER (ORDER BY wp.week_commencing)   AS row_seq
        FROM genealogy.ref_week_plan wp
        WHERE wp.week_commencing >= (SELECT dt FROM this_sunday)
          AND NOT wp.skipped
    )
    SELECT
        week_commencing,
        is_current_week,
        CAST(row_seq - 1 AS INT)          AS week_offset,
        week_index,
        COALESCE(mode_override,   mode)   AS effective_mode,
        COALESCE(branch_override, branch) AS effective_branch,
        skipped,
        notes
    FROM current_and_future
    WHERE row_seq <= 4
    ORDER BY week_commencing
"""))
```

## Cell 4 — Skip a week

Mark a week as skipped. The app advances past it and shows the following week as current. The UI **Skip Week** button calls `/week/skip` which runs the same UPDATE automatically — use this cell only for manual skips from Databricks.

```python
# Cell 4: Skip a specific week
# The app UI "Skip Week" button calls /week/skip which runs the same UPDATE automatically.
# Edit week_to_skip then run this cell for manual skips.

week_to_skip = 'YYYY-MM-DD'    # <-- replace with Sunday date e.g. '2026-04-05'
skip_notes   = 'Holiday week'  # <-- optional note; set to None to leave blank

notes_sql = ("'" + skip_notes + "'") if skip_notes else 'NULL'

spark.sql("""
    UPDATE genealogy.ref_week_plan
    SET    skipped    = true,
           notes      = {notes},
           created_at = current_timestamp()
    WHERE  week_commencing = DATE '{week}'
""".format(notes=notes_sql, week=week_to_skip))

print("Week %s marked as skipped." % week_to_skip)
print("Run Cell 3 to verify.")
```

## Cell 5 — Override or reset a week

Swap mode/branch for a specific week, or reset a skip/override back to rotation defaults. Set `mode_override` / `branch_override` to `None` to clear back to the rotation value.

```python
# Cell 5: Override or reset a week
# Swap mode/branch for a specific week, or reset a skip/override back to rotation defaults.
# Set mode_override / branch_override to None to clear back to the rotation value.

# Valid modes:    'housekeeping' | 'tree_building' | 'dna' | 'story_writing'
# Valid branches: 'Ball' | 'Mattock' | 'Cuthbertson' | 'McLean' |
#                 'Pearson' | 'Palmer' | 'Easter' | 'Clifford'

week_to_edit    = 'YYYY-MM-DD'      # <-- Sunday date
mode_override   = None              # <-- e.g. 'story_writing', or None to keep default
branch_override = None              # <-- e.g. 'Cuthbertson', or None to keep default
skipped         = False             # <-- True to skip, False to un-skip
edit_notes      = None              # <-- optional note

mode_sql   = ("'" + mode_override   + "'") if mode_override   else 'NULL'
branch_sql = ("'" + branch_override + "'") if branch_override else 'NULL'
notes_sql  = ("'" + edit_notes      + "'") if edit_notes      else 'NULL'

spark.sql("""
    UPDATE genealogy.ref_week_plan
    SET    skipped         = {skip},
           mode_override   = {mode},
           branch_override = {branch},
           notes           = {notes},
           created_at      = current_timestamp()
    WHERE  week_commencing = DATE '{week}'
""".format(
    skip   = str(skipped).lower(),
    mode   = mode_sql,
    branch = branch_sql,
    notes  = notes_sql,
    week   = week_to_edit
))

print("Week %s updated." % week_to_edit)
print("  mode_override=%s  branch_override=%s  skipped=%s" % (mode_override, branch_override, skipped))
print("Run Cell 3 to verify.")
```

```python
# Cell 6: Full 96-week schedule preview
# Shows all weeks with effective mode/branch after overrides applied.
display(spark.sql("""
    SELECT
        week_commencing,
        week_index,
        LPAD(CAST((week_index % 32) + 1 AS STRING), 2, '0') || '/32' AS cycle_position,
        COALESCE(mode_override,   mode)   AS effective_mode,
        COALESCE(branch_override, branch) AS effective_branch,
        skipped,
        notes
    FROM genealogy.ref_week_plan
    ORDER BY week_commencing
"""))
```
